###### Intro to Pivot Tables: http://pbpython.com/pandas-pivot-table-explained.html

```python
pivot_table_byorg = pd.pivot_table(df6,index=["ORGANIZATION","SORT_SUBTOTAL_DESCRIPTION"],
               values=["AMOUNT"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
pivot_table_byorg
```

```python
pivot_table_byorg.query('ORGANIZATION == ["FISCAL YEAR 2017 SERGEANT AT ARMS"]')
```

```python
pivot_table_byvendor_groupbydesc = pd.pivot_table(df6,index=["SORT_SUBTOTAL_DESCRIPTION", "VENDOR_NAME"],
               values=["AMOUNT"],
               aggfunc=[np.sum],fill_value=0,margins=True,)
```

```python
pivot_table_byvendor_groupbydesc
```

```python
pivot_table_byvendor = pd.pivot_table(df6,index=["VENDOR_NAME"],
                columns=["SORT_SUBTOTAL_DESCRIPTION"],
               values=["AMOUNT"],
            aggfunc=[np.sum],fill_value=0,margins=True,)
```

```python
pivot_table_byvendor
```

```python
top_vendor = pivot_table_byvendor.sort_values("All")            
```
